```python
import pandas as pd
from tabulate import tabulate

# Charger le fichier
path = "CC GENERAL.csv"
dataset = pd.read_csv(path)

# Changer le format d'affichage des nombres
pd.set_option('display.float_format', '{:.2f}'.format)

# Afficher le résumé statistique sous forme de tableau
summary_table = tabulate(dataset.describe(), headers='keys', tablefmt='grid')

# Afficher le tableau
print("Résumé statistique : ")
print(summary_table)
```

```python
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix

names = [
    'BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES',
    'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY',
    'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS',
    'PRC_FULL_PAYMENT', 'TENURE'
]
# Sélectionner les colonnes de features
features = dataset[names]

# Créer le scatter matrix
scatter_matrix(features, alpha=0.5, figsize=(12, 12), diagonal='hist')

# Afficher le scatter matrix
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

# Charger le dataset
url = "CC GENERAL.csv"
df = pd.read_csv(url)

# Sélectionner les features
features = df.iloc[:, 1:]

# Remplacer les valeurs manquantes par la moyenne
features = features.fillna(features.mean())

# Appliquer PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)

# Appliquer t-SNE
tsne = TSNE(n_components=2, random_state=42)
tsne_result = tsne.fit_transform(features)

# Visualiser les résultats de PCA et t-SNE
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], edgecolor='k')
plt.title('PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], edgecolor='k')
plt.title('t-SNE Results')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')

plt.tight_layout()
plt.show()
```
